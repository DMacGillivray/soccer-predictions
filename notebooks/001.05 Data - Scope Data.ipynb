## Load Data

```python
league = 'english-premier-league'
# std_names_fp = REF_DATA_DIR / str(league + '_' + 'std_name_dict.pkl')
# indatabet_fn = str('indatabet-free-odds-' + league + '.csv')
# football_data_fn = str('merge-1-' + league + '.csv')
load_fn = str('merge-2-' + league + '.csv')
save_fp = SCOPED_DATA_DIR / league / str(league + '-scoped-1.csv')
```

```python
df_orig = pd.read_csv(INTERIM_DATA_DIR / 'football-data' / league / load_fn,
                 parse_dates = ['date', 'date_fdcu', 'date_ic'], index_col=None)
df_orig
df_orig.head()
```

```python
df_orig.info()
```

```python
df = df_orig.copy(deep=True)
```

```python
# Add league column
df['league'] = league
```

```python
# Initial rename to tidy up column names
df.drop(columns=['date'], inplace=True)
df.rename(columns = {'season_fdcu': 'season', 'date_fdcu': 'date',
                         'h_htgoals_ic': 'h_htgoals', 'a_htgoals_ic': 'a_htgoals'}, inplace=True)
```

```python
# Calculate Results column
conditions = [df['h_ftgoals'] > df['a_ftgoals'],
              df['h_ftgoals'] == df['a_ftgoals'],
              df['h_ftgoals'] < df['a_ftgoals']]
choices = ['hwin', 'draw', 'awin']
df['result'] = np.select(conditions, choices, default='not-played')
```

```python
# Create Implied Probability Columns and leave in odds as features

# tot includes the bookie's overround
tot = 1/df['odds_hwin_bbmean'] + 1/df['odds_draw_bbmean'] + 1/df['odds_awin_bbmean']
hwin = 1/df['odds_hwin_bbmean']/tot
hdraw = 1/df['odds_draw_bbmean']/tot
adraw = 1/df['odds_draw_bbmean']/tot
awin = 1/df['odds_awin_bbmean']/tot

df['h_impliedwinprobbbmean'] = hwin
df['h_implieddrawprobbbmean'] = hdraw
df['a_implieddrawprobbbmean'] = adraw
df['a_impliedwinprobbbmean'] = awin

# Rename and split the odds columns to become features
odds_rename = {'odds_hwin_bbmean': 'h_winoddsbbmean',
              'odds_draw_bbmean': 'h_drawoddsbbmean',
              'odds_awin_bbmean': 'a_winoddsbbmean'}

df.rename(columns = odds_rename, inplace=True)
# Add columns for away team draw
df['a_drawoddsbbmean'] = df['h_drawoddsbbmean']
```

```python
# Rename image paths to be features
imagepaths_rename = {'heatmap_path': 'h_heatmap', 'shotmap_path': 'h_shotmap'}
df.rename(columns=imagepaths_rename, inplace=True)
df['a_heatmap'] = df['h_heatmap']
df['a_shotmap'] = df['h_shotmap']
```
