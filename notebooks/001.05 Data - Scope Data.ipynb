## Load Data

```python
league = 'english-premier-league'
# std_names_fp = REF_DATA_DIR / str(league + '_' + 'std_name_dict.pkl')
# indatabet_fn = str('indatabet-free-odds-' + league + '.csv')
# football_data_fn = str('merge-1-' + league + '.csv')
load_fn = str('merge-2-' + league + '.csv')
save_fp = SCOPED_DATA_DIR / league / str(league + '-scoped-1.csv')
```

```python
df_orig = pd.read_csv(INTERIM_DATA_DIR / 'football-data' / league / load_fn,
                 parse_dates = ['date', 'date_fdcu', 'date_ic'], index_col=None)
df_orig
df_orig.head()
```

```python
df_orig.info()
```

```python
df_cut = df_orig.copy(deep=True)

keeper_cols = ['season_fdcu', 'date_fdcu', 'h' , 'a', 'h_ftgoals', 'a_ftgoals',
               'h_htgoals_ic', 'a_htgoals_ic',
               'h_corners', 'h_fouls', 'h_ycards', 'h_rcards', 'h_shots', 'h_sot',
               'a_corners', 'a_fouls', 'a_ycards', 'a_rcards', 'a_shots', 'a_sot',
               'odds_hwin_bbmean', 'odds_draw_bbmean', 'odds_awin_bbmean',
               'odds_hwin_bbmax', 'odds_draw_bbmax', 'odds_awin_bbmax',
               'odds_hwin_WH', 'odds_draw_WH', 'odds_awin_WH',
               'odds_ftgoalsu2.5_bbmax', 'odds_ftgoalso2.5_bbmax',
               'odds_ftgoalsu2.5_bbmean', 'odds_ftgoalso2.5_bbmean',
              'heatmap_path', 'shotmap_path']

df_cut = df_cut[keeper_cols]
df_cut.rename(columns = {'season_fdcu': 'season', 'date_fdcu': 'date',
                         'h_htgoals_ic': 'h_htgoals', 'a_htgoals_ic': 'a_htgoals'}, inplace=True)
df_cut.head()
```

```python
df_cut.info()
```

```python
df_cut.to_csv(save_fp, index=False)
```
